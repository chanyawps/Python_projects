```python
%matplotlib inline
#sets the backend of matplotlib to the 'inline' backend
import numpy as np
#from scipy import misc
#misc.imread is deprecated in new version of scipy use imageio instead
import matplotlib.pyplot as plt
#from skimage import data
import imageio
```

```python
Astrophoto = imageio.imread('/content/shuffled_puzzle.jpg')
#It is a subclass of a numpy array
type(Astrophoto)
```

```python
#For your info, turning to ndarray
np.asarray(Astrophoto)
```

```python
Astrophoto.shape
```

```python
Astrophoto.ndim
```

```python
#show Astro photo
plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(Astrophoto)
```

```python
# Split Row from full image by 3 column - column 1
new_Apic = np.hsplit(Astrophoto, 3)
newpic_c1 = new_Apic[0]
newpic_c2 = new_Apic[1]
newpic_c3 = new_Apic[2]
plt.imshow(newpic_c1)
```

```python
# Split Row from full image by 3 column - column 2
plt.imshow(newpic_c2)
```

```python
# Split Row from full image by 3 column - column 3
plt.imshow(newpic_c3)
```

```python
# Split Column from full image by 3 rows - row 1
new_Cpic1 = np.vsplit(newpic_c1,3)
newpic_r11 = new_Cpic1[0]
newpic_r12 = new_Cpic1[1]
newpic_r13 = new_Cpic1[2]
plt.imshow(newpic_r11)
```

```python
# Split Column from full image by 3 rows - row 2
new_Cpic2 = np.vsplit(newpic_c2,3)
newpic_r21 = new_Cpic2[0]
newpic_r22 = new_Cpic2[1]
newpic_r23 = new_Cpic2[2]
plt.imshow(newpic_r21)
```

```python
# Split Column from full image by 3 rows - row 3
new_Cpic3 = np.vsplit(newpic_c3,3)
newpic_r31 = new_Cpic3[0]
newpic_r32 = new_Cpic3[1]
newpic_r33 = new_Cpic3[2]
plt.imshow(newpic_r31)
```

```python
# Final pic column 1
Final_pic1 = np.concatenate((newpic_r32,newpic_r21,newpic_r22), axis=0)
plt.imshow(Final_pic1)
```

```python
# Final pic column 2
Final_pic2 = np.concatenate((newpic_r13,newpic_r33,newpic_r11), axis=0)
plt.imshow(Final_pic2)
```

```python
# Final pic column 3
Final_pic3 = np.concatenate((newpic_r12,newpic_r23,newpic_r31), axis=0)
plt.imshow(Final_pic3)
```

```python
# Complete Astro photo
Astro_photo = np.concatenate((Final_pic1,Final_pic2,Final_pic3), axis=1)
plt.imshow(Astro_photo)
```
